```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
data = sns.load_dataset("iris")
```

```python
data.head()
```

```python
# Datasets come with Seaborn
sns.get_dataset_names()
```

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
```

```python
# draw lineplot 
sns.lineplot(x="sepal_length", y="sepal_width", data=data) 
  
# setting the title using Matplotlib
plt.title('Title using Matplotlib Function')

plt.xlim(4, 8)
  
plt.show()
```

```python
# Changing Figure Aesthetic
# draw lineplot 
sns.lineplot(x="sepal_length", y="sepal_width", data=data) 
  
# changing the theme to dark (darkgrid/whitegrid/dark/white/ticks)
sns.set_style("darkgrid")
plt.show()
```

```python
# Removal of Spines
# draw lineplot 
sns.lineplot(x="sepal_length", y="sepal_width", data=data) 
  
# Removing the spines
sns.despine()
plt.show()
```

```python
# Changing the figure Size
# changing the figure size
plt.figure(figsize = (10, 6))
  
# draw lineplot 
sns.lineplot(x="sepal_length", y="sepal_width", data=data) 
  
# Removing the spines
sns.despine()
  
plt.show()
```

```python
# Scaling the plots (paper/talk/poster)
# draw lineplot 
sns.lineplot(x="sepal_length", y="sepal_width", data=data) 
  
# Setting the scale of the plot
sns.set_context("talk")
  
plt.show()
```

```python
# Bar Plot

# loading dataset
data = sns.load_dataset("iris")
  
sns.barplot(x='species', y='sepal_length', data=data)
plt.show()
```

```python
# Count Plot

# loading dataset
data = sns.load_dataset("iris")
  
sns.countplot(x='species', data=data)
plt.show()
```

```python
# Box Plot

# loading dataset
data = sns.load_dataset("iris")
  
sns.boxplot(x='species', y='sepal_width', data=data)
plt.show()
```

```python
# Violinplot

# loading dataset
data = sns.load_dataset("iris")
  
sns.violinplot(x='species', y='sepal_width', data=data)
plt.show()
```

```python
# Stripplot

# loading dataset
data = sns.load_dataset("iris")
  
sns.stripplot(x='species', y='sepal_width', data=data)
plt.show()
```

```python
# Histogram

# loading dataset
data = sns.load_dataset("iris")
  
sns.histplot(x='species', y='sepal_width', data=data)
plt.show()
```

```python
# Distplot
# loading dataset
data = sns.load_dataset("iris")
  
sns.distplot(data['sepal_width'])
plt.show()
```

```python
# Pairplot - distribution of a single variable and relationship between two varaibles
# loading dataset
data = sns.load_dataset("iris")
  
sns.pairplot(data=data, hue="species", markers=["o", "s", "D"], corner=True )
plt.show()
```

```python
# Regression Plots

# loading dataset
data = sns.load_dataset("tips")
  
sns.lmplot(x='total_bill', y='tip', data=data)
plt.show()
```

```python
# loading dataset
data = sns.load_dataset("tips")
  
sns.regplot(x='total_bill', y='tip', data=data)
plt.show()
```

```python
# Heatmap
# loading dataset
data = sns.load_dataset("tips")
  
# correlation between the different parameters 
tc = data.corr()
  
sns.heatmap(tc)
plt.show()
```
